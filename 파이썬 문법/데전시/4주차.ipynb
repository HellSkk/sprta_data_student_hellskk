# 2. 데이터 시각화 (Matplotlib)

## 02. Matplotlib 알아보기

### 2-1. pandas & matplotlib 불러오기 


```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [ i for i in range(1, 6)],
    'B': [ i for i in range(5, 0, -1)]
})

df
```

```python
x = [ i for i in range(1, 6)]
y = [ i for i in range(2, 11, 2)]


plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example')
```

```python
df.plot(x='A', y='B')
plt.show()
```

## 3-2. 스타일 설정하기

```python
df.plot(x='A', y='B', color='green', linestyle='--', marker='o')
plt.show()
```

## 3-3. 범례 추가하기

```python
df.plot(x='A', y='B', color='red', linestyle='--', marker='o', label='Data Series')
plt.show()
```

```python
ax = df.plot(x='A', y='B', color='red', linestyle='--', marker='o')
ax.legend(['Data Series'])
plt.show()
```

## 3-4. 축, 제목 입력하기

```python
ax = df.plot(x='A', y='B', color='red', linestyle='--', marker='o')
ax.legend(['Data Series'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Title')
plt.show()
```

```python
ax = df.plot(x='A', y='B', color='red', linestyle='--', marker='o')
ax.legend(['Data Series'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Title')
ax.text(4, 3, 'Some text', fontsize=12)
ax.text(2, 2, 'Some text 2', fontsize=10)
plt.show()
```

## 3-6. 기타

```python
plt.figure(figsize=(8, 6))

x = [ i for i in range(1, 6)]
y = [ i for i in range(2, 11, 2)]

plt.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax = df.plot(x='A', y='B', color='red', linestyle='--', marker='o', ax=ax)
ax.legend(['Data Series'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Title')
ax.text(4, 3, 'Some text', fontsize=12)
ax.text(2, 2, 'Some text 2', fontsize=10)
plt.show()
```

# 04. 그래프 그리기 - 차트

## 4-1. Line

```python
import seaborn as sns

data = sns.load_dataset('flights')
```

```python
data
```

```python
data_grouped = data[['year', 'passengers']].groupby('year').sum().reset_index()
data_grouped
```

```python
plt.plot(data_grouped['year'], data_grouped['passengers'], color='blue', linestyle='--', marker='o')
plt.xlabel('year')
plt.ylabel('passangers')
plt.show()
```

## 4-2. Bar vs Histogram

```python
df = pd.DataFrame({
    '도시': ['서울', '부산', '대구', '인천'],
    '인구': [990, 250, 250, 290]
})
df
```

```python
plt.bar(df['도시'], df['인구'])
plt.xlabel('도시')
plt.ylabel('인구')
plt.title('도시 별 인구 수 ')
plt.show()
```

```python
import numpy as np
```

```python
data = np.random.randn(1000)
data.shape
```

```python
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

## 4-3. Pie

```python
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```

## 4-4. Box plot

```python
iris = sns.load_dataset("iris")
iris
```

```python
species = iris['species'].unique()
species
```

```python
sepal_lengths_list = [iris[iris['species'] == s]['sepal_length'].tolist() for s in species]
```

```python
len(sepal_lengths_list[2])
```

```python
plt.boxplot(sepal_lengths_list, labels=species)
plt.show()
```
