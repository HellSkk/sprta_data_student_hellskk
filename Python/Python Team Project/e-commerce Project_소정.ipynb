```python
# --- 2. 데이터 병합 (조인) ---

# transaction과 article 테이블을 article_id를 기준으로 병합
# 이렇게 하면 각 트랜잭션에 상품 정보가 추가됨.
merged = pd.merge(transactions, articles, on='article_id', how='left')
#print("\n--- transactions + articles 병합 후 ---")
print(merged.head())

```

```python
# 위에서 병합된 데이터와 customer 테이블을 customer_id를 기준으로 병합
# 이제 각 트랜잭션에 고객의 패션 뉴스 구독 여부가 포함됨.
final = pd.merge(merged, customer, on='customer_id', how='left')
#print("\n--- 최종 병합 데이터프레임 (transactions + articles + customer) ---")
print(final.head())
#print("\nfinal_df.info():")
final.info()
```

**3-1. 구독 여부별 총 구매 건수 상관 (X).** - 구독하지 않는 구매건수가 더 많음.

```python
# --- 3. 패션 뉴스 구독 여부별 구매 현황 분석 ---

# 3-1. 구독 여부별 총 구매 건수
purchase_counts = final.groupby('fashion_news_frequency').size().reset_index(name='total_purchases')
#print("\n--- 패션 뉴스 구독 여부별 총 구매 건수 ---")
print(purchase_counts) id 기간
```

**3-2. 구독 여부별 총 구매 금액 상관 (X).** -구독하지 않는 총 구매액이 더 크다.

```python
 #3-2. 구독 여부별 총 구매 금액
purchase_amounts = final.groupby('fashion_news_frequency')['price'].sum().reset_index(name='total_purchase_amount')
#print("\n--- 패션 뉴스 구독 여부별 총 구매 금액 ---")
print(round(purchase_amounts,2))
```

**3-3. 구독 여부별 평균 구매 금액 상관(X)** -구독하지 않는 평균 구매금액이 더 크다.

```python
# 3-3. (선택 사항) 구독 여부별 평균 구매 금액
average_purchase_amount = final.groupby('fashion_news_frequency')['price'].mean().reset_index(name='average_purchase_amount_per_item')
#print("\n--- 패션 뉴스 구독 여부별 평균 구매 금액 (건당) ---")
print(round(average_purchase_amount,4))
```
