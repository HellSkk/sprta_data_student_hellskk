```python
import numpy as np
import pandas as pd

# 1) 가상 데이터 생성
#   - 일부 값들을 np.nan으로 지정해 결측값을 만듭니다.
data = {
    'A': [1, 2, np.nan, 4, 5, np.nan, 7],
    'B': [5, 4, 2, np.nan, np.nan, 3, 1],
    'C': [2, np.nan, np.nan, 6, 7, 8, 9]
}
df = pd.DataFrame(data)
df
```

```python
# 2) 결측치 제거
df_drop = df.dropna()
df_drop
```

```python
# 3) 평균값으로 대체
df_mean = df.copy()
df_mean = df_mean.fillna(df_mean.mean(numeric_only=True))
df_mean
```

```python
# 4) 중앙값으로 대체
# 극단적인 값이 있을 때는 중앙값 사용
df_median = df.copy()
df_median = df_median.fillna(df_median.median(numeric_only=True))
df_median
```

```python
# 5) 최빈값으로 대체
df_mode = df.copy()
print(df_mode.mode())
mode_values = df_mode.mode().iloc[0]
df_mode = df_mode.fillna(mode_values)
df_mode
```
